```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

Base de dados

```python
import pickle
with open('risco_credito.pkl', 'rb') as file:
    X_credit_risk, y_credit_risk = pickle.load(file)
```

```python
y_credit_risk
```

```python
X_credit_risk
```

```python
risk_credit_tree = DecisionTreeClassifier(criterion='entropy')
risk_credit_tree.fit(X_credit_risk, y_credit_risk)
```

```python
risk_credit_tree.feature_importances_
```

```python
from sklearn import tree
previsores = ['historia', 'divida', 'garantias', 'renda']
figura, axes = plt.subplots(nrows=1,ncols=1, figsize=(10,10), dpi=300)
tree.plot_tree(risk_credit_tree, feature_names=previsores, class_names= risk_credit_tree.classes_, filled=True, fontsize=10)
```

```python
# HistÃ³ria boa, dÃ­vida alta, garantias nenhuma, renda > 35
# HistÃ³rua ruim, dÃ­vida alta, garantias adequadas, renda < 15
previsoes = risk_credit_tree.predict([[0, 0, 1, 2], [2, 0, 0, 0]])
previsoes
```
