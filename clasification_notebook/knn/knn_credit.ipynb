# Knn ALgorithm - k-Nearest Neighbors
## Dados de avaliação de crédito

É extremamente recomendado para algoritmos de KNN que os dados sejam padronizados, ou seja, que todos os valores sejam convertidos para números de uma mesma escala
 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import pickle
with open('../data/credit.pkl', 'rb') as f :
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
#Tamanho de registros para treinamento
X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
#Tamanho de registros para teste
X_credit_teste.shape, y_credit_teste.shape
```

```python
knn_credit = KNeighborsClassifier(n_neighbors=5, metric='minkowski')
knn_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = knn_credit.predict(X_credit_teste)
previsoes
```

```python
y_credit_teste
```

```python
#Avalia a porcentagem de acertos da previsão
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste, previsoes)
```

```python
#Matriz de confusão para visualização dos registros
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(knn_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes))
```
