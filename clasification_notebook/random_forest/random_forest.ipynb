```python
from sklearn.ensemble import RandomForestClassifier
```

Base de dados crédito


```python
import pickle

with open('credit.pkl', 'rb') as f:
    X_credit_train, y_credit_train, X_credit_test, y_credit_test = pickle.load(f)
    

X_credit_train.shape, y_credit_train.shape
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
random_forest_credit = RandomForestClassifier(n_estimators = 57, criterion = 'entropy', random_state = 0)
random_forest_credit.fit(X_credit_train, y_credit_train)
```

```python
previsoes = random_forest_credit.predict(X_credit_test)
previsoes
```

```python
y_credit_test
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_credit_test, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(random_forest_credit)
cm.fit(X_credit_train, y_credit_train)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, previsoes))
```
