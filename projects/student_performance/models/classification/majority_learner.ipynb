### Majority Learner
Applying Majority Learner to the classified database to identify baseline \
Value achieved: 69%

```python
import Orange
import Orange.evaluation
```

```python
base_students = Orange.data.Table('../../data/StudentsPerformance_processed.csv')
```

```python
base_students.domain
```

```python
majority = Orange.classification.MajorityLearner()
```

```python
predictions = Orange.evaluation.testing.TestOnTestData(base_students, base_students, [majority])
```

```python
Orange.evaluation.CA(predictions)
```
