### Naive Bayes
Utilizado base de dados processada com label encoder para target classes e OHE para atributos. No atributo average score, foram escalonados os valores após a aplicação do OHE. \
\
Score alcançado: 93%

```python
import sklearn.naive_bayes as nb
import pickle

from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import classification_report

```

```python
with open('../../data/students_performance.pkl', 'rb') as f: 
    X_students_train, y_students_train, X_students_test, y_students_test = pickle.load(f)
    
```

```python
X_students_train.shape, y_students_train.shape
```

```python
X_students_test.shape, y_students_test.shape
```

```python
naive_bayes = nb.GaussianNB()
naive_bayes.fit(X_students_train, y_students_train)
```

```python
predictions = naive_bayes.predict(X_students_test)
predictions
```

```python
y_students_test
```

#### Matriz de Confusão

Classe 0 → alunos do tipo A \
Classe 1 → alunos do tipo B \
Classe 2 → alunos do tipo C \

Previu corretamente 23 valores da classe A, errando 4 \
Previu corretamente 100 valores da classe B, erando 7 \
Previu corretamente 16 valores da classe C, errando 0

```python
cm = ConfusionMatrix(naive_bayes)
cm.fit(X_students_train, y_students_train)
cm.score(X_students_test, y_students_test)
```

```python
print(classification_report(y_students_test, predictions))
```
